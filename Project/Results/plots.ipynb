```python
import csv
import matplotlib.pyplot as plt
import numpy as np

mylist1 = []
mylist2 = []
with open('cpu_64_64', 'r') as read_obj:
    csv_reader = csv.reader(read_obj)
    list_cpu_64 = list(csv_reader)
    
    for i in list_cpu_64[0]:
        mylist1.append(float(i))
    
    read_obj.close()
    
with open('cpu_64_64_numpy', 'r') as read_obj:
    csv_reader = csv.reader(read_obj)
    list_cpu_64_np = list(csv_reader)
    
    for i in list_cpu_64_np[0]:
        mylist2.append(float(i))
    
    read_obj.close()

print(mylist2)
plt.xlabel("X-axis")
plt.ylabel("CPU%")
plt.title("CPU Usage Graph for 64x64 Matrix Multiplication")
plt.plot(mylist1)
plt.plot(mylist2, 'rx')
plt.legend(['Strassens', 'NumPy'])
plt.show()
```

```python
mylist1 = []
mylist2 = []
with open('cpu_256_256', 'r') as read_obj:
    csv_reader = csv.reader(read_obj)
    list_cpu_256 = list(csv_reader)
    
    for i in list_cpu_256[0]:
        num = float(i)
        if(num > 2000.0):
            continue
        mylist1.append(num)
        
    read_obj.close()
    
with open('cpu_256_256_numpy', 'r') as read_obj:
    csv_reader = csv.reader(read_obj)
    list_cpu_256_np = list(csv_reader)
    
    for i in list_cpu_256_np[0]:
        num = float(i)
        if(num > 3500.0):
            continue
        mylist2.append(num)
        if(len(mylist2) > len(mylist1)):
            break
    
    read_obj.close()

plt.xlabel("X-axis")
plt.ylabel("CPU%")
plt.title("CPU Usage Graph for 256x256 Matrix Multiplication")
plt.plot(mylist1)
plt.plot(mylist2)
plt.legend(['Strassens', 'NumPy'])
plt.show()
```
